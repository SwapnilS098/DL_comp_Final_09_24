```python
import torch
from compressai.zoo import bmshj2018_factorized_relu

# Check and set the device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# Define parameters
quality = 8
metric = "mse"  # or "ms-ssim"
model_name = "bmshj2018_factorized_relu"

# Create the model instance
net = bmshj2018_factorized_relu(quality=quality, metric=metric, pretrained=True).eval().to(device)

# Save the entire model (architecture + weights)
path = r"C:\Swapnil\Narrowband_DRONE\Image_compression_code\FLOPS\Pytorch_model_export\bmshj.pth"
torch.save(net.state_dict(), path)  # Save only the state_dict

# Load the entire model
# Recreate the model instance to load the state_dict
loaded_model = bmshj2018_factorized_relu(quality=quality, metric=metric, pretrained=False).eval().to(device)
loaded_model.load_state_dict(torch.load(path))
print("The model is loaded successfully using the .pth version")


```

```python
# Example input tensor
example_input = torch.randn(1, 3, 1232, 1640, device=device)

# Trace the model
try:
    model_traced = torch.jit.trace(net, example_input)
    traced_path = r"C:\Swapnil\Narrowband_DRONE\Image_compression_code\FLOPS\Pytorch_model_export\bmshj_traced.pt"
    model_traced.save(traced_path)
    print("Model exported to TorchScript (traced) and saved successfully")
except Exception as e:
    print(f"Error during tracing: {e}")
```

```python
# Path to the saved TorchScript model
model_path = r"C:\Swapnil\Narrowband_DRONE\Image_compression_code\FLOPS\Pytorch_model_export\bmshj_traced.pt"

# Load the TorchScript model
try:
    model = torch.jit.load(model_path)
    print("TorchScript model loaded successfully")
except Exception as e:
    print(f"Error loading model: {e}")

# Set the model to evaluation mode
model.eval()

# Example input tensor (make sure it matches the input size the model expects)
example_input = torch.randn(1, 3, 64, 64)  # Adjust the shape according to your model

# Perform inference
try:
    with torch.no_grad():  # Disable gradient calculation
        output = model(example_input)
    print("Inference completed successfully")
    print("Output:", output)
except Exception as e:
    print(f"Error during inference: {e}")
```
